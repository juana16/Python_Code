For the analysis below, use Data I-B. You can also use Data I-B along with Data I-A if required. Also, perform the analysis only for the duration : 2014-15.

Plot the GDP per capita for all the states.

```python
import glob
#read all state csv file names into a list
list_of_states=glob.glob('NAD-*.csv')
```

```python
#create a list of dataframes, each data frame has data for a state
list_of_df=[]
for file in list_of_states:
    print("reading file {0}".format(file))
    df_temp=pd.read_csv(file,header='infer',index_col='Item',encoding='unicode escape')
    list_of_df.append(df_temp)
```

```python
#function to extract state name from file name
state_from_file=lambda x: x.split('-')[1]

#concat of small df's into big df and use state names as index
df_big=pd.concat(list_of_df,keys=list(map(state_from_file,list_of_states)))

df_big
```

```python
#analysis for 14-15 only.
df_big=df_big['2014-15']
df_big.rename('States')
```

```python
plt.figure(3,figsize=(15,15))
per_cap_gsdp=df_big.loc[:,'Per Capita GSDP (Rs.)'].sort_values(axis=0)
sns.barplot(y=per_cap_gsdp.index, x=per_cap_gsdp)
plt.show()
```

Plot the percentage contribution of primary, secondary and tertiary sectors as a percentage of total GDP for all the states.

```python
plt.figure(4,figsize=(15,15))
name_of_states=list(map(state_from_file,list_of_states))
for i in range(0,len(name_of_states)):
    tot=df_big[name_of_states[i],'Gross State Domestic Product']
    prim=df_big[name_of_states[i],'Primary']
    sec=df_big[name_of_states[i],'Secondary']
    ter=df_big[name_of_states[i],'Tertiary']
    temp=[ (prim/tot*100), (sec/tot*100), (ter/tot*100) ]
    plt.subplot(4,7,i+1)
    plt.title(name_of_states[i])
    plt.bar(x=['Pri','Sec','Ter'],height=temp,color=['r','b','g'],align='center')
plt.show()
```

Categorise the states into four categories based on GDP per capita (C1, C2, C3, C4 - C1 would have the highest per capita GDP, C4 the lowest).

The quantile values are (0.20,0.5, 0.85, 1), i.e. the states lying between the 85th and the 100th percentile are in C1, those between 50th and 85th percentile are in C2 and so on.
Note: Categorisation into four categories will simplify the subsequent analysis, since comparing data of all the states would become quite exhaustive

```python
#Get the quartile values in a Series
per_cap_gsdp_q_val=per_cap_gsdp.quantile([0.20,0.5,0.85,1])

#Get the values in an array
arr=per_cap_gsdp_q_val.values

#Define the categorization function
def assign_cat(val):
    if val<=arr[0]:
        return 'C4'
    elif ( val>arr[0] and val<=arr[1]):
        return 'C3'
    elif ( val>arr[1] and val<=arr[2]):
        return 'C2'
    elif ( val>arr[2] and val<=arr[3]):
        return 'C1'    
#Apply the function on the series to get the category series
category=per_cap_gsdp.apply(assign_cat)
category
```

For each category C1, C2, C3, C4:
Find the top 3/4/5 sub-sectors (such as agriculture, forestry and fishing, crops, manufacturing etc.) [not primary, secondary and tertiary] which contribute to approx. 80% of the GSDP of each category

Note-I: The nomenclature for this project is as follows: primary, secondary and tertiary are named 'sectors', while agriculture, manufacturing etc. are named 'sub-sectors'

Note-II: If the top-3 sub-sectors contribute to say 79% of the GDP of some category, you can report 'these top-3 sub-sectors contribute to approx. 80% of the GDP'. This is to simplify the analysis and make the results consumable (remember, the CEO has to present the report to the CMs, and CMs have limited time, so the analysis needs to be very sharp and concise)

```python
#Rename the index at level 0
df_big.index.rename('States',level=0,inplace=True)
#Rename the category Series index
category.index.rename('States',inplace=True)
category.rename('Categories',inplace=True)

#Merge the two data frames to get the data frame containing Item, Value, and Category with States
df_big_copy=df_big.reset_index()
df_with_cat= pd.merge(df_big_copy,category.reset_index(),how='inner',left_on='States',right_on='States')

#set the index to categories and then to items
df_with_cat.reset_index()
df_with_cat.set_index(['Categories','Item'],inplace=True)

#drop all variables that are not needed in the analysis
df_relevant=df_with_cat.drop(labels=['Primary','Secondary','Tertiary','TOTAL GSVA at basic prices','Taxes on Products','Subsidies on products','Per Capita GSDP (Rs.)'],level=1)

#combine the sub sector data for all states by using indexing with sum
df_combined_sub_sector=df_relevant.sum(axis=0,level=[0,1])

#now that the sum values have been obtained we need to obtain the GDP for each category
#to do that we need the data frame with lookup over items
df_clean=df_combined_sub_sector.reset_index()
df_cleaner=df_clean.set_index('Item')

#this dataframe holds the total gsdp per category
total_gsdp=df_cleaner.loc['Gross State Domestic Product']

#differentiate the column name so that adding values to larger data frame is cleaner
total_gsdp=total_gsdp.rename(columns={'2014-15':'GSDP_Total'})

#add the total gdp per category column to the df containing sub sector gdp per category
df_perc=pd.merge(df_clean,total_gsdp,how='inner',left_on='Categories',right_on='Categories')
#add the percentage column
df_perc['percen']=df_perc['2014-15']/df_perc['GSDP_Total']*100

#set the index on the sub sectors
df_perc=df_perc.set_index('Item')

#drop the gsdp per category and the total population per category since we are interested in 
df_perc_1=df_perc.drop(labels=['Gross State Domestic Product','Population (\'00)'],axis=0)

#remove unimportant columns for cleanliness
df_perc_1=df_perc_1.drop(columns=['2014-15','GSDP_Total'])

#after all cleanup is done resent the index so that further analysis is straightforward
df_perc_1=df_perc_1.reset_index()
```

Plot the contribution of the sub-sectors as a percentage of the GSDP of each category. 

```python
df_sorted=df_perc_1.sort_values(by=['Categories','percen'])
```

```python
plt.figure(5,figsize=(16,20))
for i in range(1,5):
    plt.subplot(4,1,i)
    category='C'+str(i)
    df_temp=df_sorted[df_sorted['Categories']==category]
    sns.barplot(x=df_temp['percen'],y=df_temp['Item'],)
    plt.title(category)
    plt.xlabel('Percentage')
    plt.ylabel('Category')
plt.show()
```
